```python
import pandas as pd #create frame to work with data
import numpy as np #calc
import matplotlib.pyplot as plt #plot
from astropy.table import Table #imports table
from astropy.io import fits #imports fits file


Obv1080Table=Table.read("26965_200205.1080.fits")
Obv1110Table=Table.read("26965_201127.1110.fits")

Obv1080Table.info ()
Obv1110Table.info ()
```

```python
Obv1080fits=fits.open("26965_200205.1080.fits")
Obv1110fits=fits.open("26965_201127.1110.fits")
Obv1080fits.info ()
Obv1110fits.info ()
```

# 1080

```python
#This is extracted values graph
wavelength1080 = Obv1080Table["wavelength"][85]

plt.plot(Obv1080Table['continuum'][85],Obv1080Table['spectrum'][85])
plt.plot(Obv1080Table['continuum'][85],Obv1080Table['continuum'][85])

plt.title('Extracted Values')
plt.xlabel('Wavelength[A]')
plt.ylabel('Extracted Counts')

plt.show()
```

```python
#Spectrum with Blaze
spectrumblaze1080 = Obv1080Table['spectrum'][85]*Obv1080Table['blaze'][85]
blazecontinuum1080 = Obv1080Table['blaze'][85]/Obv1080Table['continuum'][85]

plt.plot(wavelength1080,spectrumblaze1080)
plt.plot(wavelength1080,blazecontinuum1080*1000) #Multiplied by 1000 because I was experimenting with the graph dialation

plt.title('Spectrum with Blaze')
plt.xlabel('Wavelength[A]')
plt.ylabel('Raw Counts')

plt.show()
```

```python
plt.plot(wavelength1080,blazecontinuum1080)
```

```python
#Continuum Normalized Spectrum
spectrumcontinuum1080 = Obv1080Table['spectrum'][85]/Obv1080Table['continuum'][85]
plt.plot(wavelength1080,spectrumcontinuum1080)
plt.plot(wavelength1080, Obv1080Table['continuum'][85]/Obv1080Table['continuum'][85])

plt.title('Continuum Normalized Spectrum')
plt.xlabel('Wavelength[A]')
plt.ylabel('Continuum Normalized Counts')

plt.show()
```
