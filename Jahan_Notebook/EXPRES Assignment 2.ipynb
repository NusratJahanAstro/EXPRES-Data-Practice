```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

file_1080 = '26965_200205.1080.fits'
file_1110 = '26965_201127.1110.fits'

hdus = fits.open(file_1080)
hdus1 = fits.open(file_1110)
```

```python
hdus.info()

hdus1.info()
```

```python
wave = hdus[1].data['wavelength'].copy()
cont = hdus[1].data['continuum'].copy()
spec = hdus[1].data['spectrum'].copy()
unct = hdus[1].data['uncertainty'].copy()
hdus.close()
```

```python
wave1 = hdus1[1].data['wavelength'].copy()
cont1 = hdus1[1].data['continuum'].copy()
spec1 = hdus1[1].data['spectrum'].copy()
unct1 = hdus1[1].data['uncertainty'].copy()
hdus1.close()
```

```python
val = input("Enter your order number: ")
nord = int(val)

plt.plot(wave[nord], spec[nord]/cont[nord])
plt.plot(wave1[nord], spec1[nord]/cont1[nord],c='m')

plt.plot(wave[nord],cont[nord]/cont[nord])
plt.plot(wave1[nord],cont1[nord]/cont1[nord])

plt.title('Continuum Normalized Spectrum')
plt.xlabel('Wavelength[A]')
plt.ylabel('Continuum Normalized Counts')

plt.show()

#when comparing the two orders from either sightings, there are slight similarities and differences.
```

```python
for order in range(0,85):
    plt.plot(wave[order], spec[order]/cont[order])
plt.title('Continuum Normalized Spectrum for 1080')
plt.xlabel('Wavelength[A]')
plt.ylabel('Continuum Normalized Counts')
plt.show()
```

```python
for order2 in range(0,85):
    plt.plot(wave1[order2], spec1[order2]/cont1[order2])
plt.title('Continuum Normalized Spectrum for 1110')
plt.xlabel('Wavelength[A]')
plt.ylabel('Continuum Normalized Counts')
plt.show()
```
